# Creando una tortuga

Una imagen vale más que mil palabras, por eso vamos a descubrir este genial módulo sobre la marcha mediante distintos ejemplo.

Empecemos por lo más básico, crear una instancia de la tortuga mediante la clase `Turtle`:

```python
from mobilechelonian import Turtle

kame = Turtle()
```

Una tortuga es una pluma para dibujar gráficos primitivos, formas geométricas básicas. 

¿Pero por qué una tortuga? Imaginad este animal en medio de la arena, si se mueve dejará un rastro en la arena parecido a un dibujo, solo puede moverse hacia adelante y para cambiar la dirección debe rotar sobre sí misma.

Podemos consultar la documentación con todas las instrucciones posibles con:

```python
help(Turtle)
```

Así resumidos son los siguientes:

* `backward`: Mueve la tortuga hacia atrás una distancia.
* `circle`: Dibuja un círculo o parte de un círculo.
* `forward`: Mueve la tortuga hacia adelante una distancia.
* `home`: Mueve la tortuga a la posición inicial.
* `left`: Gira la tortuga una cantidad de grados a la izquierda.
* `pencolor`: Cambia el color de la pluma, por defecto es negro.
* `pendown`: Baja la pluma, se inicia con la pluma bajada. 
* `penup`: Sube la pluma haciendo que no se pueda dibujar.
* `right`: Gira la tortuga una cantidad de grados a la derecha.
* `setbearing`: Cambia los grados de rotación de la dirección.
* `setposition`: Establece la posición `x` e `y` sobre el canvas.
* `speed`: Cambia la velocidad de la tortuga entre 1 y 10.

Con ellos podemos realizar todo tipo de dibujos.

## Moviendo la tortuga

Vamos a crear de nuevo la tortuga y le diremos que se mueva hacia adelante una distancia (en píxeles), de manera que deje un rastro sobre el lienzo:

```python
kame = Turtle()

kame.forward(100)
```

Podemos encadenar diferentes instrucciones, de manera que la tortuga gire sobre sí misma y se vuelva a mover:

```python
kame = Turtle()

# Adelante 100 pixels
kame.forward(100)
# Rotamos 90º a la izquierda
kame.left(90)
# Adelante 100 pixels
kame.forward(100)
# Volvemos al principio
kame.home()
```

Veamos donde se posiciona la tortuga si le indicamos que se dirija a la coordenada `(0,0)`:

```python
kame = Turtle()

# Movemos la tortuga a la coordenada (0,0)
kame.setposition(0, 0)
```

Como vemos `(0,0)` es la esquina superior izquierda.

El lienzo mide `400x400`, con eso en mente podemos decirle que dibuje un cuadrado mediante las coordenadas de sus vértices:

```python
kame = Turtle()

# Incrementamos la velocidad
kame.speed(5)
# Levantamos la pluma para no dibujar
kame.penup()
# Movemos la tortuga al primer vértice
kame.setposition(100, 100)
# Bajamos la pluma para dibujar
kame.pendown()
# Movemos la tortuga a los cuatro vértices
kame.setposition(300, 100)
kame.setposition(300, 300)
kame.setposition(100, 300)
kame.setposition(100, 100)
```

También podemos dibujar círculos con un radio, no sé por qué pero me los dibuja muy lento aunque incrementa la velocidad:

```python
kame = Turtle()
kame.speed(10)

kame.circle(20)
kame.circle(40)
kame.circle(80)
```

Para cambiar la posición inicial hay que hacerlo con `setposition`:

```python
kame = Turtle()
kame.speed(10)

kame.penup()
kame.setposition(200, 210)
kame.pendown()
kame.circle(20)

kame.penup()
kame.setposition(200, 230)
kame.pendown()
kame.circle(40)

kame.penup()
kame.setposition(200, 280)
kame.pendown()
kame.circle(80)
```
